```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img = cv2.imread('../CourseFiles/DATA/00-puppy.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

```python
img_gray = cv2.imread('../CourseFiles/DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
np.shape(img_gray)
```

```python
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
new_img = cv2.resize(fixed_img, (960,720))
plt.imshow(new_img)
```

```python
flipx_img = cv2.flip(fixed_img,0)
plt.imshow(flipx_img)
```

```python
flipy_img = cv2.flip(fixed_img,1)
plt.imshow(flipy_img)
```

```python
cv2.imwrite('./flipped_img.jpg', flipy_img)
```

```python
flipy_fixed_img = cv2.cvtColor(flipy_img, cv2.COLOR_BGR2RGB)
plt.imshow(flipy_fixed_img)
```

```python
cv2.imwrite('./flipped__fixed_img.jpg', flipy_fixed_img)
```
